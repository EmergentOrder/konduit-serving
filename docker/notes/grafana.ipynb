Grafana
--------------------------------


[Grafana](https://grafana.com) is a dashboard system for pulling data from different sources and displaying it in real time.
In our self contained examples, we use it to visualize output from prometheus.

The grafana system in the containers can be accessed from: http://localhost:3000

The user/password are set in the docker-compose.yml. The defaults as of this writing are: admin/password

Update the docker-compose.yml if you want new credentials.
